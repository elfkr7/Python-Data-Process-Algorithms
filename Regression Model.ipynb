## Q3. Train-Test split 

### 3.1 split the data into training and test sets. Test set should cover 20% of the original data records
* **Make sure to define `random_state` as 123 just so that the split can be reproduced**

```python

#### START YOUR CODE HERE ####

## hint: import the scikit learn library on train-test split
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(
    data, test_size=0.2, random_state=123
)
####  END YOUR CODE HERE  ####
```

### 3.2 split the data into training and test sets via stratified sampling on variable SEX 
* **Make sure to define `random_state` as 123 just so that the split can be reproduced**

```python

#### START YOUR CODE HERE ####

## hint: check the method we used in our lecture
from sklearn.model_selection import StratifiedShuffleSplit
sss = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=123)
for train_index, test_index in sss.split(data, data["SEX"]):
    strat_train_set = data.loc[train_index]
    strat_test_set = data.loc[test_index]

####  END YOUR CODE HERE  ####
```

### 3.3 check if the stratified sampling helps on making the SEX distribution balanced between train and test sets 

```python

#### START YOUR CODE HERE ####
compare_props = pd.DataFrame({
    "Overall": data.SEX.value_counts()/data.shape[0],
    "Stratified": strat_test_set["SEX"].value_counts()/len(strat_test_set),
    "Random": test_set["SEX"].value_counts()/len(test_set),
}).sort_index()
compare_props["Rand. %error"] = 100 * compare_props["Random"]/compare_props["Overall"] - 100
compare_props["Strat. %error"] = 100 * compare_props["Stratified"]/compare_props["Overall"] - 100
compare_props
####  END YOUR CODE HERE  ####
```

## Q4. Model Fit and Evaluation

### 4.1 Fit a Linear Regression model on the training set

```python

#### START YOUR CODE HERE ####
## hint 1: import the scikit learn library on linear regression
from sklearn.linear_model import LinearRegression
## hint 2: create the linear regression model object
lin_reg = LinearRegression()
## hint 3: fit the model on the training set
lin_reg.fit(X_train, y_train)
####  END YOUR CODE HERE  ####
```

### 4.2 Get the intercept and coefficients

```python
#### START YOUR CODE HERE ####

intercept = lin_reg.intercept_[0]
coefs = lin_reg.coef_[0]


print('The intercept value is {}'.format(intercept))
####  END YOUR CODE HERE  ####
```

Create a dataframe, in which the 1st column is the variable name and the 2nd column is the corresponding fitting coefficient

```python
#### START YOUR CODE HERE ####
df_feature_importance = pd.DataFrame({
    'attributes':list(X_train.columns),
    'importance':coefs
}).sort_values(by='importance',ascending=False).reset_index(drop=True)
df_feature_importance

####  END YOUR CODE HERE  ####
```

### Get the RMSE and $R^2$ values in training set 

```python
#### START YOUR CODE HERE ####

from sklearn.metrics import mean_squared_error


y_train_pred = lin_reg.predict(X_train)
rmse_train =  np.sqrt(mean_squared_error(y_train, y_train_pred))
r2_train = lin_reg.score(X_train, y_train)

print('RMSE: {}'.format(rmse_train))
print('R2: {}'.format(r2_train))
####  END YOUR CODE HERE  ####
```

### Get the RMSE and $R^2$ values in test set

```python
#### START YOUR CODE HERE ####

y_test_pred = lin_reg.predict(X_test)
rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))
r2_test = lin_reg.score(X_test, y_test)

print('RMSE: {}'.format(rmse_test))
print('R2: {}'.format(r2_test))
####  END YOUR CODE HERE  ####
```
