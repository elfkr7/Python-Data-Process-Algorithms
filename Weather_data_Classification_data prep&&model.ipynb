```python
import numpy as np # linear algebra
import pandas as pd # data processing
import matplotlib.pyplot as plt # plot
# import seaborn as sns
```

```python
# Read the dataset
df = pd.read_csv("weatherAUS.csv")
df.head()
```

# Data Preprocessing 

```python
# Initial exploration
df.describe().T
```

```python
df.info()
```

```python

pd.options.display.max_rows=None
pd.options.display.max_columns=None
dfXX=pd.DataFrame(None,index=[df.columns])
dfXX['dtype'] = dict(df.dtypes).values()
dfXX.T
```

```python
# Calculate the percentages of missing values
null_percentages = df.isnull().sum() / len(df)
null_percentages
```

```python
# Get rid of the columns which have more than 35% missing data
null_features = null_percentages[null_percentages > 0.35].index
df_dropped = df.drop(null_features, axis = 1)
df_dropped.head()
```

```python
# Get rid of the pointless column
df_dropped = df_dropped.drop(['Date'], axis = 1)
df_dropped.head()
```

```python
# Get rid of the null rows
df_dropped = df_dropped.dropna()
df_dropped.head()
```

```python
# Recheck the number of count and null for 'df_dropped'
count_num = pd.DataFrame(df_dropped.count())
null_num = pd.DataFrame(df_dropped.isnull().sum())
df_dropped_describe = pd.concat([count_num, null_num], axis = 1)
df_dropped_describe.columns = ['count', 'null']
df_dropped_describe['dtype'] = dict(df_dropped.dtypes).values()
df_dropped_describe
```

```python
# Group the data by numerical and categorical variables
categorical_var = df_dropped[list(df_dropped_describe.loc[df_dropped_describe['dtype']=='object'].index)]
numerical_var = df_dropped[list(df_dropped_describe.loc[df_dropped_describe['dtype']=='float64'].index)]
```

```python
# Recheck the number of count for 'numerical_var'
count_num_1 = pd.DataFrame(numerical_var.count())
count_num_1.columns = ['count']
count_num_1['dtype'] = dict(numerical_var.dtypes).values()
count_num_1
```

```python
# Recheck the number of count for 'categorical_var'
count_num_1 = pd.DataFrame(categorical_var.count())
count_num_1.columns = ['count']
count_num_1['dtype'] = dict(categorical_var.dtypes).values()
count_num_1
```

```python
# Check the numerical variables
numerical_var.head()
```

```python
# Check the categorical variables
categorical_var.head()
```

```python
# Scale the numerical variables
from sklearn.preprocessing import StandardScaler
numerical_var = StandardScaler().fit_transform(numerical_var)
numerical_var.shape
```

```python
# Transform the categorical variables into numbers
from sklearn.preprocessing import LabelEncoder
Arrays = []
for col in list(categorical_var):
    Encoder = LabelEncoder()
    Encoder.fit(categorical_var[col])
    Arrays.append(Encoder.transform(categorical_var[col]))
categorical_var = Arrays[0]
for i in range(len(Arrays)-1):
    categorical_var = np.vstack((categorical_var, Arrays[i+1]))
categorical_var = categorical_var.T
categorical_var.shape
```

```python
# Regroup the data by explanatory variables and response variable
X_all = np.hstack((categorical_var[:,0:-1],numerical_var))
y = categorical_var[:,-1]
```

```python
X_all.shape
```

```python
y.shape
```

```python
# Check the correlation of columns
c = pd.DataFrame(X_all, columns = list(df_dropped)[0:-1]).corr()
c
```

```python
# Find the pairs of columns which have a strong correlation (more than 0.95) with each other
row_num_and_col_num = []
row_name_and_col_name = []
c = np.array(c)
for i in range(len(c)):
    for j in range(i+1, len(c)):
        if abs(c[i,j]) >= 0.95:
            row_num_and_col_num.append((i,j))
            row_name_and_col_name.append((list(df_dropped)[i],list(df_dropped)[j]))
row_name_and_col_name
```

```python
# Get rid of the columns which have a strong correlation with others
cols = list(range(0, row_num_and_col_num[0][0])) + list(range((row_num_and_col_num[0][0]+1), row_num_and_col_num[1][1])) + list(range((row_num_and_col_num[1][1]+1), len(c)))
X = X_all[:,cols]
X.shape
```

# Models

```python
# Split the dataset to train set and test set
from sklearn.model_selection import train_test_split
train, test, train_labels, test_labels = train_test_split(X,
                                                          y,
                                                          test_size=0.33,
                                                          random_state=42)
```

```python
# import all the models
from sklearn.naive_bayes import GaussianNB
from sklearn import neighbors
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
methods = [GaussianNB(), neighbors.KNeighborsClassifier(), tree.DecisionTreeClassifier(criterion='gini'), RandomForestClassifier(), GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, max_depth=1, random_state=0)]
methods_names = ['Naive-Bayes classifier', 'k-nearest neighbors algorithm', 'decision trees', 'random forests', 'GBDT']

# run all the models in a loop
for i in range(len(methods)):
    
    # model initialization
    model = methods[i]

    # model fitting & training
    model.fit(train, train_labels)

    # model prediction
    preds = model.predict(test)

    # model evaluation
    print('This is the result of ', methods_names[i], ':', sep="")
    print('The accuracy rate of is ', 100 * accuracy_score(test_labels, preds), '%!', sep="")
    print('The precision rate of is ', 100 * precision_score(test_labels, preds), '%!', sep="")
    print('The recall rate of is ', 100 * recall_score(test_labels, preds), '%!', sep="")
    print('The f1 score of is ', 100 * f1_score(test_labels, preds), '%!', sep="")
    print('----------------------------------------------------------------------')
```
